```python
fruits = [
    "apple", "banana", "orange", "grape", "strawberry",
    "watermelon", "korean melon", "peach", "plum", "kiwi",
    "mango", "pineapple", "cherry"
]

for fruit in fruits:
  print(fruit)
```

```python
upper_fruits = []

for fruit in fruits:
    upper_fruits.append(fruit.upper())

print(upper_fruits)
```

```python
long_fruits = []

for fruit in fruits:
    if len(fruit) >= 6:
        long_fruits.append(fruit)

print(long_fruits)
```

```python
count = 1
for fruit in fruits:
  print(f"{count}. {fruit}")
  count += 1
```

```python
count = 1
kings = [
    "태조", "정종", "태종", "세종", "문종",
    "단종", "세조", "예종", "성종", "연산군",
    "중종", "인종", "명종", "선조", "광해군",
    "인조", "효종", "현종", "숙종", "경종",
    "영조", "정조", "순조", "헌종", "철종",
    "고종", "순종"
]

for king in kings:
  print(f"{count}대왕 {king}")
  count +=1
```

```python
n = int(input("조선 몇 대 왕의 이름을 알고 싶은가요?:"))

if 1<= n <= len(kings):
  king_name= kings[n-1]
  print(f"{n}대 왕은 {king_name}입니다.")

else:
  print("1부터 27 사이의 숫자를 입력하세요")
```

```python
name = input("찾고 싶은 왕의 이름은?: ")

i = 0
while i <len(kings):
  if kings[i] == name:
    print(f"{name}는 조선 제 {i + 1}대 왕입니다.")
    break
  i += 1
```

```python
name = input("찾고 싶은 왕의 이름은?: ")

if name in kings:
  order = kings.index(name) +1
  print(f"{name}은 조선 제 {order}대 왕입니다.")
else:
  print(f"{name}은 조선 왕 계보에 없습니다.")
```
