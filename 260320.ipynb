```python
# 산지니
minutes = int(input("분을 입력하세요: "))

hours = minutes // 60
mins = minutes % 60

print(f"{minutes}분은 {hours}시간 {mins}분입니다.")
```

```python
# 사용자에게 정수를 입력받는다
num = float(input("숫자를 입력하세요: "))

# 2로 나눈 나머지(%)가 0이면 짝수, 아니면 홀수
if num > 0:
    print("양수입니다.")
elif num < 0:
    print("음수입니다.")
else:
    print("0입니다")
```

```python
coffee = int(input('커피의 개수를 입력하시오: '))
latte  = int(input('라떼의 개수를 입력하시오: '))
green_tea = int(input('녹차의 개수를 입력하시오: '))
juice =  int(input('주스의 개수를 입력하시오: '))

coffee_price = int (input('커피의 가격을 입력하시오: '))
latte_price = int (input('라떼의 가격을 입력하시오: '))
green_tea_price = int (input('녹차의 가격을 입력하시오: '))
juice_price = int (input('주스의 가격을 입력하시오: '))

sum = coffee*coffee_price+latte*latte_price+green_tea*green_tea_price+juice*juice_price
print(f'총 주문 금액은 {sum}원입니다')
```
