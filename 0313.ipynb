```python
# 사칙연산 프로그램 (변수 + 연산자만)

a = 12                 # [1] 계산에 쓸 첫 번째 숫자를 변수에 저장해요. 나중에 a만 바꾸면 전체 결과가 같이 바뀌어요.
b = 5                  # [2] 계산에 쓸 두 번째 숫자를 변수에 저장해요.

add_result = a + b     # [3] 덧셈 결과를 변수에 저장해요. 바로 출력해도 되지만, 저장해두면 재사용/확인이 쉬워요.
sub_result = a - b     # [4] 뺄셈 결과를 변수에 저장해요.
mul_result = a * b     # [5] 곱셈 결과를 변수에 저장해요.
div_result = a / b     # [6] 나눗셈 결과를 변수에 저장해요. /는 보통 소수점이 나올 수 있어요.

print("a =", a)        # [7] 어떤 값으로 계산했는지 보여주기 위해 출력해요(디버깅/확인에 도움).
print("b =", b)        # [8] 두 번째 값도 같이 출력해요.

print("a + b =", add_result)  # [9] 덧셈 결과를 화면에 보여줘요.
print("a - b =", sub_result)  # [10] 뺄셈 결과를 보여줘요.
print("a * b =", mul_result)  # [11] 곱셈 결과를 보여줘요.
print("a / b =", div_result)  # [12] 나눗셈 결과를 보여줘요.
```

```python
# 사칙연산 프로그램 (입력 버전)

a_text = input("첫 번째 숫자를 입력하세요: ")   # [1] 사용자에게 값을 입력받아요. input() 결과는 '글자'예요.
b_text = input("두 번째 숫자를 입력하세요: ")  # [2] 두 번째 값도 같은 방식으로 입력받아요.

a = int(a_text)                                # [3] 글자로 들어온 값을 정수 숫자로 바꿔요. 그래야 + - * / 계산이 돼요.
b = int(b_text)                                # [4] 두 번째 입력도 정수로 바꿔요.

add_result = a + b                             # [5] 덧셈
sub_result = a - b                             # [6] 뺄셈
mul_result = a * b                             # [7] 곱셈
div_result = a / b                             # [8] 나눗셈(/은 결과가 소수일 수 있어요)

print("a =", a)                                # [9] 실제로 어떤 숫자로 계산했는지 확인용 출력
print("b =", b)                                # [10] 확인용 출력

print("a + b =", add_result)                   # [11] 덧셈 결과 출력
print("a - b =", sub_result)                   # [12] 뺄셈 결과 출력
print("a * b =", mul_result)                   # [13] 곱셈 결과 출력
print("a / b =", div_result)                   # [14] 나눗셈 결과 출력
```
