```python
import random
answer=int(random.randint(1,100))
count=1

while True:
  num=int(input("값을 입력하세요: "))
  if num==answer:
    print(f"정답입니다! {count}번째에 맞혔습니다.")
    break
  elif num > answer:
    print("정답보다 큰 수입니다. 좀더 작은 수를 생각해보세요.")
    count+=1
  else:
    print("정답보다 작습니다. 좀더 큰 수를 생각해보세요.")
    count+=1
```
