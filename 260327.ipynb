```python
import random

user1_raw = random.randint(1,3)
user2_raw = random.randint(1,3)

if user1_raw == 1:
  user1 = "가위"
elif user1_raw == 2:
  user1 = "바위"
else:
  user1 = "보"

if user2_raw == 1:
  user2 = "가위"
elif user2_raw == 2:
  user2 = "바위"
else:
  user2 = "보"

print(f"user1:{user1}")
print(f"user2:{user2}")

if user1_raw == user2_raw:
  print("무승부입니다.")

elif user1_raw == 1 and user2_raw == 3:
  print("user1이 이겼습니다.")

elif user1_raw == 2 and user2_raw == 1:
  print("user1이 이겼습니다.")

elif user1_raw == 3 and user2_raw == 2:
  print("user1이 이겼습니다.")

else:
  print("user2가 이겼습니다.")
```
