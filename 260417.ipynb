```python
cafe_menu = {
    "아메리카노": 4500,
    "카페라떼": 5500,
    "바닐라라떼": 6000,
    "카푸치노": 5500,
    "카페모카": 6000,
    "카라멜마키아토": 6500,
    "에스프레소": 4000,
    "콜드브루": 5800,
    "디카페인 아메리카노": 5000,
    "레몬에이드": 6500,
    "자몽에이드": 6800,
    "초코라떼": 6000,
    "말차라떼": 6500,
    "밀크티": 6500,
    "티(캐모마일)": 5500,
}

for key, value in cafe_menu.items():
    print(f"{key}의 가격은 {value}원입니다")
```

```python
order_counts = {}
total_price = 0

while True:
  print("\n====메뉴판====")
  for name, price in cafe_menu.items():
    print(f"{name} : {price}원")
  print("=================")

  menu_name = input("주문(메뉴 입력) 또는 종료: ")
  if menu_name == "종료":
    break
  if menu_name not in cafe_menu:
    print("없는 메뉴입니다.")
    continue
  qty_str= input("개수를 입력하세요: ").strip()

  if not qty_str.isdigit() or int(qty_str) <= 0:
    print("개수는 1이상의 정수로 입력하세요")
    continue

  qty = int(qty_str)
  if menu_name in order_counts:
    order_counts[menu_name] = order_counts[menu_name] + qty
  else:
    order_counts[menu_name] = qty
  total_price += cafe_menu[menu_name] * qty

  print(f"추가 완료: {menu_name} * {qty} (현재 합계: {total_price})")
```

```python
menu_sum = 0
for value in cafe_menu.values():
  menu_sum += value
print(f"총 메뉴 가격은 {menu_sum}원입니다.")
```

```python
total_drinks = int(input("주문할 음료 개수를 입력하세요: "))

sum = 0
for i in range(total_drinks):
  menu = input("음료명을 입력하세요: ")
  if menu in cafe_menu:
    sum += cafe_menu[menu]
  else:
    print(f"{menu}는 메뉴에 없습니다.")
print(f"주문하신 총 금액은 {sum}원입니다.")
```
